```python
import random

class Game:
    """Classe représentant un jeu de Pierre-Feuille-Ciseaux."""

    def __init__(self):
        """Constructeur de la classe Game"""
        self.choices = ["rock", "paper", "scissors"]

    def get_user_item(self):
        while True:
            print("Choisissez votre arme :")
            print("   • rock (pierre)     → tapez 'rock'")
            print("   • paper (feuille)   → tapez  'paper'")
            print("   • scissors (ciseaux) → tapez  'scissors'")

            user_input = input("\nVotre choix : ").strip().lower()

            # Vérification si l'utilisateur a tapé le mot complet
            if user_input in self.choices:
                print(f"\nVous avez choisi : {user_input.upper()} !")
                return user_input

            # Si le choix est invalide
            print("\nChoix invalide ! Veuillez choisir parmi : rock, paper, scissors")

    def get_computer_item(self):
        computer_choice = random.choice(self.choices)
        print(f"\nL'ordinateur a choisi : {computer_choice.upper()} !")
        return computer_choice

    def get_game_result(self, user_item, computer_item):

        # Règles du jeu
        rules = {
            ("rock", "scissors"): "win",
            ("scissors", "paper"): "win",
            ("paper", "rock"): "win",
            ("rock", "paper"): "loss",
            ("paper", "scissors"): "loss",
            ("scissors", "rock"): "loss"
        }

        # Égalité
        if user_item == computer_item:
            return "draw"

        # Victoire ou défaite
        return rules[(user_item, computer_item)]

    def play(self):

        # Étape 1 : Obtenir le choix de l'utilisateur
        user_item = self.get_user_item()

        # Étape 2 : Obtenir le choix de l'ordinateur
        computer_item = self.get_computer_item()

        # Étape 3 : Déterminer le résultat
        result = self.get_game_result(user_item, computer_item)

        # Étape 4 : Afficher le résultat de manière conviviale

        # messages selon le résultat
        if result == "win":
            print(f"VOUS AVEZ GAGNÉ !")
            print(f"   {user_item} bat {computer_item} → Félicitations !")
        elif result == "loss":
            print(f"VOUS AVEZ PERDU ! ")
            print(f"   {computer_item} bat {user_item} → Meilleure chance la prochaine fois !")
        else:
            print(f"ÉGALITÉ !")
            print(f"   Vous avez tous les deux choisi {user_item} → Match nul !")


        return result
```

```python
from game import Game

def get_user_menu_choice():
    print("MENU PRINCIPAL")
    print("=" * 50)
    print("1. Jouer une nouvelle partie")
    print("2. Voir les scores")
    print("3. Quitter")

    while True:
        choice = input("\nQue voulez-vous faire ? (1/2/3) : ").strip()
        if choice in ['1', '2', '3']:
            return choice
        else:
            print("Choix invalide ! Veuillez entrer 1, 2 ou 3.")

def print_results(results):

    print("STATISTIQUES FINALES ")
    print("=" * 50)

    total_games = sum(results.values())

    # Affichage des résultats avec emojis
    print(f"\nNombre total de parties : {total_games}")
    print("DÉTAIL DES RÉSULTATS :")

    print(f"Victoires   : {results['win']}")
    print(f"Défaites    : {results['loss']}")
    print(f"Égalités    : {results['draw']}")

def main():
    print("   BIENVENUE DANS LE JEU PIERRE-FEUILLE-CISEAUX !   ")

    # Initialisation du dictionnaire des résultats
    results = {
        'win': 0,
        'loss': 0,
        'draw': 0
    }

    # Boucle principale du jeu
    while True:
        choice = get_user_menu_choice()

        if choice == '1':  # Jouer une nouvelle partie

            print("   LANCEMENT D'UNE NOUVELLE PARTIE   ")


            # Création et exécution d'une partie
            game = Game()
            game_result = game.play()

            # Mise à jour des statistiques
            results[game_result] += 1

            # Message de confirmation
            result_fr = {
                'win': 'gagné ',
                'loss': 'perdu ',
                'draw': 'fait match nul '
            }
            print(f"\nPartie enregistrée : VOUS AVEZ {result_fr[game_result].upper()} !")

        elif choice == '2':  # Voir les scores
            print("\nAFFICHAGE DES SCORES ACTUELS")
            total = sum(results.values())
            if total == 0:
                print("\nAucune partie n'a encore été jouée.")
                print("Jouez d'abord une partie pour voir vos scores !")
            else:
                print(f"\nStatistiques depuis le début de la session :")
                print(f"   • Victoires : {results['win']}")
                print(f"   • Défaites  : {results['loss']}")
                print(f"   • Égalités  : {results['draw']}")
                print(f"   • Total     : {total}")


        elif choice == '3':  # Quitter
            print("\nFERMETURE DU PROGRAMME... ")
            # Affichage du résumé final avant de quitter
            print_results(results)
            break

    print("\nAu revoir et à bientôt pour de nouvelles parties !")

# Point d'entrée du programme
if __name__ == "__main__":
    main()
```
