```python
import math
import turtle
from ColabTurtlePlus.Turtle import *
```

```python
class Circle:
    """
    Classe qui représente un cercle.
    """

    def __init__(self, radius):
        if radius <= 0:
            raise ValueError("Le rayon doit être strictement positif.")

        self._radius = float(radius)

    @property
    def radius(self):
        return self._radius

    @radius.setter
    def radius(self, value):
        if value <= 0:
            raise ValueError("Le rayon doit être strictement positif.")

        self._radius = float(value)

    @property
    def diameter(self):
        return self._radius * 2

    @diameter.setter
    def diameter(self, value):
        if value <= 0:
            raise ValueError("Le diamètre doit être strictement positif.")

        self._radius = value / 2

    @property
    def area(self):
        return math.pi * (self.radius ** 2)

    def __str__(self):
        return (
            f"Cercle(Rayon = {self.radius:.2f}, "
            f"Diamètre = {self.diameter:.2f}, "
            f"Aire = {self.area:.2f})"
        )

    def __add__(self, other):
        if not isinstance(other, Circle):
            return NotImplemented

        return Circle(self.radius + other.radius)

    def __gt__(self, other):
        if not isinstance(other, Circle):
            return NotImplemented

        return self.radius > other.radius

    def __lt__(self, other):
        if not isinstance(other, Circle):
            return NotImplemented

        return self.radius < other.radius

    def __eq__(self, other):
        if not isinstance(other, Circle):
            return NotImplemented

        return self.radius == other.radius

    def afficher_separateur(titre):
        print("=" * 60)
        print(f" {titre} ".center(70, "="))
        print("=" * 60)
```

```python
afficher_separateur("CRÉATION DES CERCLES")
cercle1 = Circle(4)
cercle2 = Circle(7)

print(cercle1)
print(cercle2)

print("\n")

# =====================================================
# Utilisation du diamètre
# =====================================================

afficher_separateur("MODIFICATION PAR LE DIAMÈTRE")

cercle1.diameter = 20

print(cercle1)

print("\n")

# =====================================================
# Calcul de l'aire
# =====================================================

print("=" * 60)
print("        CALCUL DE L'AIRE")
print("=" * 60)

print(f"Aire du cercle 1 : {cercle1.area:.2f}")
print(f"Aire du cercle 2 : {cercle2.area:.2f}")

print("\n")

# =====================================================
# Addition
# =====================================================

print("=" * 60)
print("        ADDITION DE DEUX CERCLES")
print("=" * 60)

nouveau_cercle = cercle1 + cercle2

print(nouveau_cercle)

print("\n")

# =====================================================
# Comparaisons
# =====================================================

print("=" * 60)
print("        COMPARAISONS")
print("=" * 60)

print("Cercle1 > Cercle2 :", cercle1 > cercle2)
print("Cercle1 == Cercle2 :", cercle1 == cercle2)

print("\n")

# =====================================================
# Tri des cercles
# =====================================================

print("=" * 60)
print("        TRI DES CERCLES")
print("=" * 60)

liste = [
    Circle(12),
    Circle(2),
    Circle(7),
    Circle(5),
    Circle(9),
    Circle(1)
]

liste.sort()

for cercle in liste:
    print(cercle)
```

```python
clearscreen()
setup(250, 80)
t = Turtle()
t.speed(10) # Vitesse rapide

# Position de départ
x_pos = -100
y_pos = 0

for cercle in liste:
    rayon = cercle.radius

    # Se positionner
    t.penup()
    t.goto(x_pos, y_pos - rayon)
    t.pendown()

    # Dessiner le cercle
    t.circle(rayon)

    # Écrire le rayon au centre
    t.penup()
    t.goto(x_pos, y_pos - 10)
    t.write(f"{int(rayon)}", align="center")

    # Décalage pour le prochain cercle
    x_pos += (rayon * 2) + 20
```
