# Exercice 1

```python
# Classes de base fournies
class Pets():
    def __init__(self, animals):
        self.animals = animals

    def walk(self):
        for animal in self.animals:
            print(animal.walk())

class Cat():
    is_lazy = True
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def walk(self):
        return f'{self.name} is just walking around'

class Bengal(Cat):
    def sing(self, sounds):
        return f'{sounds}'

class Chartreux(Cat):
    def sing(self, sounds):
        return f'{sounds}'

# ÉTAPE 1 : Création de la classe Siamese qui hérite de Cat
class Siamese(Cat):
    def __init__(self, name, age):
        # Appel du constructeur de la classe parente Cat
        super().__init__(name, age)

    def walk(self):
        # On peut personnaliser la méthode walk pour les Siamois si nécessaire
        return f'{self.name} le Siamois  se promène élégamment'

    def meow(self):
        return f'{self.name} fait un miaulement distinctif de Siamois !'

# ÉTAPE 2 : Création d'une liste d'instances de chats
# Création des objets chats avec différents noms et âges
bengal_obj = Bengal("Tigrou", 3)
chartreux_obj = Chartreux("Bleuet", 2)
siamese_obj = Siamese("Sakura", 4)

all_cats = [bengal_obj, chartreux_obj, siamese_obj]

# ÉTAPE 3 : Création d'une instance de Pets avec la liste de chats
sara_pets = Pets(all_cats)


# ÉTAPE 4 : Promener tous les chats
print("Promenade des chats de Sara \n")
sara_pets.walk()


# Bonus
print("\n Fonctionnalités bonus \n")

# Test des méthodes sing pour Bengal et Chartreux
print(bengal_obj.sing("Miaou Miaou !"))
print(chartreux_obj.sing("Ronron..."))

# Test de la méthode spécifique au Siamese
print(siamese_obj.meow())

# Affichage des attributs du Siamese
print(f"Âge de {siamese_obj.name} : {siamese_obj.age} ans")
print(f"Les chats sont-ils paresseux ? {siamese_obj.is_lazy}")
```

# Exercice 2

```python
class Dog:
    def __init__(self, name, age, weight):
        self.name = name
        self.age = age
        self.weight = weight

    def bark(self):
        return f"{self.name} is barking"

    def run_speed(self):
        return (self.weight / self.age) * 10

    def fight(self, other_dog):
        # Calcul de la puissance de combat pour chaque chien
        self_power = self.run_speed() * self.weight
        other_power = other_dog.run_speed() * other_dog.weight

        # Détermination du gagnant
        if self_power > other_power:
            return f"{self.name} a gagné le combat contre {other_dog.name} !"
        elif other_power > self_power:
            return f"{other_dog.name} a gagné le combat contre {self.name} !"
        else:
            return f"Match nul entre {self.name} et {other_dog.name} !"


# ÉTAPE 2 : Création de trois instances de chiens
print(" Création des chiens \n")

# Chien 1 : Un berger allemand puissant mais âgé
rex = Dog("Rex", 8, 35)

# Chien 2 : Un jeune et rapide husky
luna = Dog("Luna", 2, 22)

# Chien 3 : Un petit mais costaud bouledogue
max = Dog("Max", 4, 28)


# ÉTAPE 3 : Test des méthodes
print("\n Test des méthodes \n")

# Test de la méthode bark()
print("Test des aboiements ")
print(rex.bark())
print(luna.bark())
print(max.bark())

print("\n Test des vitesses de course ")
# Test de la méthode run_speed()
print(f"Vitesse de {rex.name} : {rex.run_speed():.2f} unités")
print(f"Vitesse de {luna.name} : {luna.run_speed():.2f} unités")
print(f"Vitesse de {max.name} : {max.run_speed():.2f} unités")

print("\n Test des combats ")
# Test de la méthode fight()
print(rex.fight(luna))
print(rex.fight(max))
print(luna.fight(max))
```

# Exercice 3

```python
import random
from exercise_2_dog import Dog  # Import de la classe Dog depuis l'exercice précédent


class PetDog(Dog):

    def __init__(self, name, age, weight):
        # Appel du constructeur de la classe parente Dog
        super().__init__(name, age, weight)
        # Ajout de l'attribut spécifique aux chiens domestiques
        self.trained = False  # Par défaut, le chien n'est pas dressé

    def train(self):
        print(self.bark())  # Affiche l'aboiement du chien
        self.trained = True  # Le chien est maintenant dressé
        print(f"{self.name} est maintenant dressé ! ")

    def play(self, *args):
        # Extraction des noms des chiens
        dog_names = []

        for arg in args:

            if isinstance(arg, Dog):
                dog_names.append(arg.name)
            else:
                dog_names.append(str(arg))
        # Ajout du nom du chien actuel
        dog_names.append(self.name)
        # Création de la liste des noms sous forme de chaîne
        names_string = ", ".join(dog_names)
        # Affichage du message de jeu
        print(f" {names_string} jouent tous ensemble ! ")

    def do_a_trick(self):
        # Liste des tours possibles
        tricks = ["does a barrel roll", "stands on his back legs", "shakes your hand", "plays dead"]

        if self.trained:
            # Sélection aléatoire d'un tour
            chosen_trick = random.choice(tricks)
            print(f" {self.name} {chosen_trick} ! ")
        else:
            print(f" {self.name} n'est pas encore dressé. Il faut d'abord l'entraîner !")


# Création de trois instances de PetDog

my_dog = PetDog("Fido", 2, 10)
my_dog.train()
my_dog.play("Buddy", "Max")
my_dog.do_a_trick()

print("\n Création des chiens :")
rex_pet = PetDog("Rex", 5, 32)
luna_pet = PetDog("Luna", 3, 18)
max_pet = PetDog("Max", 4, 25)


print("\n Dressage de Luna ")
luna_pet.train()

print("\n Dressage de Rex ")
rex_pet.train()


print("\n Jeu entre chiens ")
luna_pet.play("Rex", "Max")

print("\n Jeu entre chiens ")
luna_pet.play(rex_pet, max_pet)

print("\n Jeu avec deux chiens seulement ")
max_pet.play("Luna")


print("\n Tours de Luna (dressée) ")
for i in range(3):  # Exécute 3 tours aléatoires
    luna_pet.do_a_trick()

print("\n Tours de Max (non dressé) ")
max_pet.do_a_trick()

print("\n Dressage de Max ")
max_pet.train()
```

# Exercice 4

```python
class Person:


    def __init__(self, first_name, age, last_name=""):

        self.first_name = first_name
        self.age = age
        self.last_name = last_name

    def is_18(self):

        return self.age >= 18

    def __str__(self):

        return f"{self.first_name} {self.last_name} ({self.age} ans)"


class Family:


    def __init__(self, last_name):

        self.last_name = last_name
        self.members = []  # Liste vide pour stocker les objets Person

    def born(self, first_name, age):

        # Création d'une nouvelle personne avec le nom de famille
        new_person = Person(first_name, age, self.last_name)

        # Ajout à la liste des membres
        self.members.append(new_person)

        # Message de bienvenue
        if age == 0:
            print(f" Bienvenue à {first_name} {self.last_name} qui vient de naître ! ")
        else:
            print(f" {first_name} {self.last_name} ({age} ans) a rejoint la famille ! ")

    def check_majority(self, first_name):

        # Recherche de la personne dans la liste des membres
        found_person = None

        for person in self.members:
            if person.first_name == first_name:
                found_person = person
                break

        # Si la personne n'existe pas dans la famille
        if found_person is None:
            print(f" Désolé, {first_name} ne fait pas partie de la famille {self.last_name}.")
            return

        # Vérification de la majorité
        if found_person.is_18():
            print(f" {found_person.first_name} {self.last_name} est majeur(e).")
            print(f"   Vous avez {found_person.age} ans, vos parents acceptent que vous sortiez avec vos amis.")
        else:
            age_restant = 18 - found_person.age
            print(f" {found_person.first_name} {self.last_name} est mineur(e) ({found_person.age} ans).")
            print(f"   Désolé, vous n'êtes pas autorisé(e) à sortir avec vos amis.")
            print(f"   Il vous reste {age_restant} an(s) avant d'être majeur(e).")

    def family_presentation(self):


        if not self.members:
            print("Cette famille n'a pas encore de membres.")
        else:
            print(f"\nNombre de membres : {len(self.members)}")
            print("\nListe des membres :")
            print("-" * 30)

            for i, member in enumerate(self.members, 1):
                statut = " Adulte" if member.is_18() else " Enfant"
                print(f"{i}. {member.first_name} {member.last_name} - {member.age} ans - {statut}")

        print("=" * 50 + "\n")

    def get_member_by_name(self, first_name):

        for person in self.members:
            if person.first_name == first_name:
                return person
        return None




# ÉTAPE 1 : Création d'une famille
famille_martin = Family("Martin")
print(f"\n Famille {famille_martin.last_name} créée avec succès !")

# ÉTAPE 2 : Ajout de membres avec la méthode born()


print("\n--- Naissance de Sophie ---")
famille_martin.born("Sophie", 25)

print("\n--- Naissance de Thomas ---")
famille_martin.born("Thomas", 17)

print("\n--- Arrivée d'Emma ---")
famille_martin.born("Emma", 12)

print("\n--- Naissance de Lucas ---")
famille_martin.born("Lucas", 30)

# ÉTAPE 3 : Présentation de la famille
famille_martin.family_presentation()

# ÉTAPE 4 : Tests de la méthode check_majority()

print("\n--- Vérification pour Sophie (25 ans) ---")
famille_martin.check_majority("Sophie")

print("\n--- Vérification pour Thomas (17 ans) ---")
famille_martin.check_majority("Thomas")

print("\n--- Vérification pour Emma (12 ans) ---")
famille_martin.check_majority("Emma")

print("\n--- Vérification pour Lucas (30 ans) ---")
famille_martin.check_majority("Lucas")

print("\n--- Vérification pour une personne inexistante ---")
famille_martin.check_majority("Pierre")

# ÉTAPE 5 : Tests supplémentaires

# Création d'une deuxième famille pour démonstration
print("\n--- Création d'une deuxième famille : Famille Dubois ---")
famille_dubois = Family("Dubois")
famille_dubois.born("Julie", 8)
famille_dubois.born("Nicolas", 19)
famille_dubois.family_presentation()

# Test de la méthode get_member_by_name
print("--- Recherche d'un membre spécifique ---")
membre = famille_martin.get_member_by_name("Thomas")
if membre:
    print(f" Membre trouvé : {membre}")
else:
    print(" Membre non trouvé")
```
