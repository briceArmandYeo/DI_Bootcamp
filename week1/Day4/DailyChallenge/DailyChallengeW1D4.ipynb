```python
import math

class Pagination:

    def __init__(self, items=None, page_size=10):
        # Initialisation de la liste des éléments
        if items is None:
            self.items = []
        else:
            self.items = items

        # Taille de chaque page
        self.page_size = page_size

        # Index de la page courante
        self.current_idx = 0

        # Calcul du nombre total de pages
        if len(self.items) == 0:
            self.total_pages = 1
        else:
            self.total_pages = math.ceil(len(self.items) / self.page_size)

    def _validate_page_number(self, page_num):

        if page_num < 1 or page_num > self.total_pages:
            raise ValueError(f"Page {page_num} invalide. Les pages vont de 1 à {self.total_pages}.")

    def get_visible_items(self):
        start = self.current_idx * self.page_size
        end = start + self.page_size
        return self.items[start:end]

    def go_to_page(self, page_num):
        self._validate_page_number(page_num)
        self.current_idx = page_num - 1  # Conversion base 1 -> base 0
        return self

    def first_page(self):
        self.current_idx = 0
        return self

    def last_page(self):
        self.current_idx = self.total_pages - 1
        return self

    def next_page(self):
        if self.current_idx < self.total_pages - 1:
            self.current_idx += 1
        return self

    def previous_page(self):
        if self.current_idx > 0:
            self.current_idx -= 1
        return self

    def __str__(self):
        visible_items = self.get_visible_items()
        if not visible_items:
            return "Aucun élément sur cette page."

        # Chaque élément sur une nouvelle ligne
        return "\n".join(str(item) for item in visible_items)


# Création de la liste alphabétique
alphabetList = list("abcdefghijklmnopqrstuvwxyz")
p = Pagination(alphabetList, 4)


print(p.get_visible_items())

p.next_page()
print(p.get_visible_items())

p.last_page()
print(p.get_visible_items())

p.go_to_page(10)
print(p.current_idx + 1)

p.go_to_page(0)
```
