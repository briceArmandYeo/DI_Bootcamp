# Challenge 1

```python
number = int(input("Entrez le nombre : "))
length = int(input("Entrez la longueur de la liste : "))

multiples = [number * i for i in range(1, length + 1)]

print(f"Liste des {length} multiples de {number} :")
print(multiples)
```

# Challenge 2

```python
word = input("Entrez une chaîne de caractères : ")
if not word:
    print("Chaîne vide.")
else:
    result = word[0]
    for i in range(1, len(word)):
        if word[i] != word[i-1]:
            result += word[i]
    print(f"Résultat : {result}")
```
