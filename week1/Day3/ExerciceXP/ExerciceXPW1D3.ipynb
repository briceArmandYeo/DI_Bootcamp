# Exercice 1

```python
class Cat:
    def __init__(self, cat_name, cat_age):
        self.name = cat_name
        self.age = cat_age

# Étape 1 : Créer trois objets Chat
cat1 = Cat("Minou", 4)
cat2 = Cat("Félix", 7)
cat3 = Cat("Gribouille", 2)

# Étape 2 : Fonction pour trouver le chat le plus vieux
def trouver_chat_le_plus_vieux(cat1, cat2, cat3):
    if cat1.age >= cat2.age and cat1.age >= cat3.age:
        return cat1
    elif cat2.age >= cat1.age and cat2.age >= cat3.age:
        return cat2
    else:
        return cat3

# Étape 3 : Trouver et afficher le chat le plus vieux
chat_vieux = trouver_chat_le_plus_vieux(cat1, cat2, cat3)

print(f"Le chat le plus vieux est {chat_vieux.name}, et il a {chat_vieux.age} ans.")
```

# Exercice 2

```python
class Dog:
    def __init__(self, name, height):
        self.name = name
        self.height = height

    def bark(self):
        print(f"{self.name} fait ouaf !")

    def jump(self):
        hauteur_saut = self.height * 2
        print(f"{self.name} saute {hauteur_saut} cm de haut !")

# Étape 2 : Créer les chiens
davids_dog = Dog("Rex", 50)
sarahs_dog = Dog("Milou", 25)

# Étape 3 : Afficher les infos et tester les méthodes
print("Le chien de David ")
print(f"Nom : {davids_dog.name}")
print(f"Taille : {davids_dog.height} cm")
davids_dog.bark()
davids_dog.jump()

print("\nLe chien de Sarah ")
print(f"Nom : {sarahs_dog.name}")
print(f"Taille : {sarahs_dog.height} cm")
sarahs_dog.bark()
sarahs_dog.jump()

# Étape 4 : Comparer les tailles
print("\nComparaison ")
if davids_dog.height > sarahs_dog.height:
    print(f"{davids_dog.name} est plus grand que {sarahs_dog.name} !")
elif davids_dog.height < sarahs_dog.height:
    print(f"{sarahs_dog.name} est plus grand que {davids_dog.name} !")
else:
    print("Les deux chiens ont la même taille !")
```

# Exercice 3

```python
class Song:
    def __init__(self, lyrics):
        self.lyrics = lyrics

    def sing_me_a_song(self):
        for ligne in self.lyrics:
            print(ligne)


# Exemple d'utilisation
stairway = Song([
    "There’s a lady who's sure",
    "all that glitters is gold",
    "and she’s buying a stairway to heaven"
])

# On chante la chanson
stairway.sing_me_a_song()
```

# Exercice 4

```python
class Zoo:
    def __init__(self, zoo_name):
        self.zoo_name = zoo_name
        self.animaux = []  # Liste pour stocker les noms des animaux

    def add_animal(self, *nouveaux_animaux):
        for animal in nouveaux_animaux:
            if animal not in self.animaux:
                self.animaux.append(animal)
                print(f"{animal} a été ajouté au zoo.")
            else:
                print(f"{animal} est déjà dans le zoo, il n'a pas été ajouté.")

    def get_animaux(self):
        if not self.animaux:
            print(f"Le zoo {self.zoo_name} est vide pour le moment.")
        else:
            print(f"Voici les animaux du zoo {self.zoo_name} :")
            for animal in self.animaux:
                print(f"  - {animal}")

    def sell_animal(self, animal_vendu):
        if animal_vendu in self.animaux:
            self.animaux.remove(animal_vendu)
            print(f"{animal_vendu} a été vendu et a quitté le zoo.")
        else:
            print(f"{animal_vendu} n'est pas dans le zoo, impossible de le vendre.")

    def sort_animaux(self):
        if not self.animaux:
            print("Aucun animal à trier.")
            return {}

        animaux_tries = sorted(self.animaux)

        # Dictionnaire pour stocker les groupes
        groupes = {}

        for animal in animaux_tries:
            premiere_lettre = animal[0].upper()
            if premiere_lettre not in groupes:
                groupes[premiere_lettre] = []
            groupes[premiere_lettre].append(animal)

        return groupes

    def get_groupes(self):
        groupes = self.sort_animaux()

        if not groupes:
            print("Le zoo est vide ou n'a pas encore été trié.")
            return

        print(f"\n Groupes d'animaux du zoo {self.zoo_name} \n")
        for lettre, liste_animaux in sorted(groupes.items()):
            print(f"{lettre} : {liste_animaux}")

mon_zoo = Zoo("Parc Animalier de la Savane")



print(" Ajout d'animaux \n")

mon_zoo.add_animal("Girafe", "Lion", "Zèbre", "Babouin", "Ours", "Lion")

print("\n Affichage des animaux \n")
mon_zoo.get_animaux()

print("\n Vente d'un animal \n")
mon_zoo.sell_animal("Ours")
mon_zoo.sell_animal("Pingouin")

print("\n Après la vente \n")
mon_zoo.get_animaux()

print("\n Groupes par première lettre \n")
mon_zoo.get_groupes()

print("\n Bonus : Ajout groupé d'animaux \n")
mon_zoo.add_animal("Éléphant", "Tigre", "Panda")

print("\n Affichage final \n")
mon_zoo.get_animaux()
mon_zoo.get_groupes()
```
