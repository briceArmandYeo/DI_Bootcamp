# Exercice 2

```python
family = {"rick": 43, 'beth': 13, 'morty': 5, 'summer': 8}
# Initialisation du coût total à 0
total_cost = 0

# Parcours de chaque membre de la famille
for name, age in family.items():
    # Détermination du prix selon l'âge
    if age < 3:
        price = 0
    elif age <= 12:  # age entre 3 et 12 inclus
        price = 10
        categorie = "Enfant (10$)"
    else:  # age > 12
        price = 15
        categorie = "Adulte (15$)"
    # Ajout du prix au coût total
    total_cost += price
    # Affichage du prix pour chaque membre
    print(f"{name:10} | Âge: {age:2} | {categorie:15} | Prix: {price:2} $")
# Affichage du total
print(f"\nCoût TOTAL pour la famille : {total_cost} $")
```

```python
# Dictionnaire vide pour stocker les membres saisis par l'utilisateur
user_family = {}
# Saisie du nombre de membres
nb_membres = int(input("Combien de membres dans la famille ? "))

for i in range(nb_membres):
    print(f"\nMembre #{i+1}")
    nom = input("Entrez le nom du membre : ")
    age = int(input(f"Entrez l'âge de {nom} : "))
    # Ajout du membre au dictionnaire
    user_family[nom] = age

total_cost_user = 0

for name, age in user_family.items():
    if age < 3:
        price = 0
    elif age <= 12:
        price = 10
        categorie = "Enfant (10$)"
    else:
        price = 15
        categorie = "Adulte (15$)"


    total_cost_user += price


    print(f"{name:10} | Âge: {age:2} | {categorie:15} | Prix: {price:2} $")

print(f"\nCoût TOTAL pour votre famille : {total_cost_user} $")
```

# Exercice 3

```python
# Création du dictionnaire brand avec toutes les informations fournies
brand = {
    "name": "Zara",
    "creation_date": 1975,
    "creator_name": "Amancio Ortega Gaona",
    "type_of_clothes": ["men", "women", "children", "home"],
    "international_competitors": ["Gap", "H&M", "Benetton"],
    "number_stores": 7000,
    "major_color": {
        "France": ["blue"],
        "Spain": ["red"],
        "US": ["pink", "green"]
    }
}

print(brand)

# Changer la valeur de number_stores à 2
brand["number_stores"] = 2

# Afficher une phrase décrivant les clients de Zara
clothes_list = brand["type_of_clothes"]
clothes_str = ", ".join(clothes_list)
print(f"Zara vend des vêtements pour : {clothes_str}.\n")

# Ajouter une nouvelle clé country_creation avec la valeur "Spain"
brand["country_creation"] = "Spain"

# Vérifier si international_competitors existe et ajouter "Desigual"
if "international_competitors" in brand:
    brand["international_competitors"].append("Desigual")
    print(f"Nouvelle liste : {brand['international_competitors']}\n")

# Supprimer la clé creation_date
if "creation_date" in brand:
    del brand["creation_date"]

# Afficher le dernier élément de international_competitors

last_competitor = brand["international_competitors"][-1]
print(f"{last_competitor}\n")

# Afficher les couleurs principales aux US

us_colors = brand["major_color"]["US"]
print(f"{us_colors}\n")

# Afficher le nombre de clés dans le dictionnaire
number_of_keys = len(brand)
print(f"{number_of_keys} clés\n")

# Afficher toutes les clés du dictionnaire
all_keys = list(brand.keys())
print(f"{all_keys}\n")


# Création d'un deuxième dictionnaire more_on_zara
more_on_zara = {
    "creation_date": 1975,
    "number_stores": 7000
}

print(more_on_zara)

brand.update(more_on_zara)
print(brand)

for key, value in brand.items():
    print(f"{key:25} : {value}")
```

# Exercie 5

```python
import random

def compare_numbers(user_number):
    random_number = random.randint(1, 100)
    if user_number == random_number:
        print("Success!")
    else:
        print(f"Fail! Your number: {user_number}, Random number: {random_number}")

print("Test avec le nombre 50 :")
compare_numbers(50)
print()


print("Test avec le nombre 42 :")
compare_numbers(42)
print()

print("Test avec le nombre 7 :")
compare_numbers(7)
print()

nombre_utilisateur = int(input("Entrez un nombre entre 1 et 100 : "))
if 1 <= nombre_utilisateur <= 100:
    compare_numbers(nombre_utilisateur)
else:
    print("Erreur : Le nombre doit être compris entre 1 et 100 !")
```

# Exercice 7

```python
import random
```

```python
def get_random_temp():
    return random.randint(-10, 40)

get_random_temp()
```

```python
def main():

    # Appel à get_random_temp()
    temperature = get_random_temp()

    # Affichage du message amical avec la température
    print(f"The temperature right now is {temperature} degrees Celsius.")

    if temperature < 0:
        return "Brrr, that's freezing! Wear some extra layers today."
    elif temperature <= 16:
        return "Quite chilly! Don't forget your coat."
    elif temperature <= 23:
        return "Nice weather."
    elif temperature <= 32:
        return "A bit warm, stay hydrated."
    else:
        return "It's really hot! Stay cool."
```

```python
main()
```

```python
def get_random_temp_float():
    return round(random.uniform(-10, 40),2)

get_random_temp_float()
```

```python
# Liste vide pour stocker tous les ingrédients choisis par le client
toppings = []

# Prix de base de la pizza
prix_base = 10

# Prix par ingrédient supplémentaire
prix_par_topping = 2.50


while True:
    # Demander à l'utilisateur d'entrer un ingrédient
    topping = input("Entrez un ingrédient (ou 'quit' pour terminer) : ")

    # Vérifier si l'utilisateur veut arrêter
    if topping.lower() == 'quit':
        break  # Sortie de la boucle

    if topping.strip():
        # Ajout de l'ingrédient à la liste
        toppings.append(topping)

        # Message de confirmation avec formatage
        print(f"Ajout de {topping} à votre pizza.")
        print()
    else:
        print("Veuillez entrer un ingrédient valide.\n")


# Vérifier si des ingrédients ont été choisis
if toppings:
    # Afficher tous les ingrédients
    print("\nVos ingrédients :")
    for i, topping in enumerate(toppings, 1):
        print(f"  {i}. {topping}")


    nombre_toppings = len(toppings)
    cout_total = prix_base + (nombre_toppings * prix_par_topping)

    # Affichage des détails du prix

    print(f"Prix de base              : {prix_base:.2f} $")
    print(f"{nombre_toppings} ingrédient(s) x {prix_par_topping:.2f}$ : {nombre_toppings * prix_par_topping:.2f} $")

    print(f"TOTAL À PAYER             : {cout_total:.2f} $")

else:
    # Cas où aucun ingrédient n'a été choisi
    print("\nVous n'avez sélectionné aucun ingrédient.")
    print(f"Pizza nature : {prix_base:.2f} $")
```
