# Challenge 1

```python
# Demander à l'utilisateur d'entrer un mot
mot = input("Entrez un mot : ")

# Créer un dictionnaire vide pour stocker les lettres et leurs indices
dictionnaire = {}

# Parcourir chaque caractère du mot avec son index
for index, lettre in enumerate(mot):

    if lettre in dictionnaire:

        dictionnaire[lettre].append(index)
    else:

        dictionnaire[lettre] = [index]


print(dictionnaire)
```

# Challenge 2

```python
# Données fournies
items_purchase = {
    "Water": "$1",
    "Bread": "$3",
    "TV": "$1,000",
    "Fertilizer": "$20"
}
wallet = "$300"

# Fonction pour nettoyer les prix (enlever $ et les virgules)
def nettoyer_prix(prix_avec_symbole):
    # Enlever le signe dollar et les virgules, puis convertir en entier
    return int(prix_avec_symbole.replace("$", "").replace(",", ""))

# Nettoyer le montant du portefeuille
argent_disponible = nettoyer_prix(wallet)

# Créer une liste pour les articles abordables
panier = []

# Parcourir les articles dans l'ordre du dictionnaire (priorité)
for article, prix_texte in items_purchase.items():
    prix_article = nettoyer_prix(prix_texte)

    if prix_article <= argent_disponible:
        panier.append(article)
        argent_disponible -= prix_article

# Afficher le résultat
if panier:
    print("You can buy:")
    panier_trie = sorted(panier)
    print(panier_trie)
else:
    print("Nothing")
```

```python
def nettoyer_prix(prix_avec_symbole):
  return int(prix_avec_symbole.replace("$", "").replace(",", ""))

# Fonction pour nettoyer les prix (enlever $ et les virgules)
def achat (items_purchase,wallet):
  # Nettoyer le montant du portefeuille
  argent_disponible = nettoyer_prix(wallet)

  # Créer une liste pour les articles abordables
  panier = []

  # Parcourir les articles dans l'ordre du dictionnaire (priorité)
  for article, prix_texte in items_purchase.items():
      prix_article = nettoyer_prix(prix_texte)

      if prix_article <= argent_disponible:
          panier.append(article)
          argent_disponible -= prix_article

  # Afficher le résultat
  if panier:
      print("You can buy:")
      panier_trie = sorted(panier)
      return print(panier_trie)
  else:
      return print("Nothing")
```

```python
items_purchase = {
    "Water": "$1",
    "Bread": "$3",
    "TV": "$1,000",
    "Fertilizer": "$20"
}
wallet = "$300"
achat(items_purchase,wallet)
```
