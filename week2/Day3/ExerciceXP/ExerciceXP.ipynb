```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

mu, sigma = 50, 10
gaussian_dist = stats.norm(mu, sigma)
print("PDF at x=0.5:", gaussian_dist.pdf(0.5))
x_gaussian = np.linspace(0, 100)
pdf_gaussian = gaussian_dist.pdf(x_gaussian)
plt.plot(x_gaussian, pdf_gaussian)
plt.title('Gaussian Distribution')
plt.xlabel('x')
plt.ylabel('PDF')
plt.show()
```

# Exercice 4

```python
import numpy as np
from scipy import stats

data1 = np.random.normal(50, 10, 100)
data2 = np.random.normal(60, 10, 100)


t_stat, p_value = stats.ttest_ind(data1, data2)

print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

alpha = 0.05
if p_value < alpha:
    print("Conclusion: Rejet de l'hypothèse nulle (les moyennes sont significativement différentes)")
else:
    print("Conclusion: Échec du rejet de l'hypothèse nulle (pas de différence significative)")
```

# Exercice 5

```python
import numpy as np
from scipy import stats

house_sizes = [50, 70, 80, 100, 120]
house_prices = [150.000, 200.000, 210.000, 250.000, 280.000]


slope, intercept, r_value, p_value, std_err = stats.linregress(house_sizes, house_prices)


print(f"la pente est : {slope:.2f}")
print(f"l'interception de la ligne de régression est : {intercept:.2f}" )
```

```python
size_to_predict = 90
predicted_price = slope * size_to_predict + intercept

print(f"la prédiction pour une maison de {size_to_predict}m² est : {predicted_price:.3f}")
```

Interpretation

Pour chaque augmentation de 1 m² de la surface de la maison, le prix estimé du logement augmente en moyenne d’environ 1,83 unités monétaires.

la relation entre la surface et le prix est donc positive : plus la maison est grande, plus son prix tend à augmenter.


# Exercice 6

```python
from scipy import stats

fertilizer_1 = [5, 6, 7, 6, 5]
fertilizer_2 = [7, 8, 7, 9, 8]
fertilizer_3 = [4, 5, 4, 3, 4]
```

```python
f_value, p_value = stats.f_oneway(fertilizer_1, fertilizer_2, fertilizer_3)
print("F-value:", f_value)
print("P-value:", p_value)
```

La P-value est 2.7e-05 soit environ 0,000027
P-value < 0,05. Cela signifie que nous rejetons l'hypothèse nulle par consequent il existe une différence statistiquement significative entre les moyennes d'au moins deux de ces engrais.

Si la P-value était supérieure à 0,05 , On ne pourrait pas rejeter l'hypothèse nulle.
On peut conclure que les différences observées entre les groupes sont probablement dues au hasard.



# Exercice 7

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

n = 10
p = 0.5
k = np.arange(0, n + 1)
binom_dist = stats.binom(n, p)
probabilities = binom_dist.pmf(k)

print(probabilities)
```

```python
k_target = 5
prob_5 = binom.pmf(k_target, n, p)

print(f"La probabilité d'obtenir exactement {k_target} têtes est : {prob_5:.2f}")
```

# Exercice 8

```python
import pandas as pd
from scipy.stats import pearsonr
```

```python
data = pd.DataFrame({'age': [23, 25, 30, 35, 40], 'income': [35000, 40000, 50000, 60000, 70000]})
```

```python
corr, _ = pearsonr(data['age'], data['income'])
print(f'Pearson Correlation Coefficient: {corr:.2f}')
```

```python
from scipy.stats import spearmanr

corr, _ = spearmanr(data['age'], data['income'])
print(f'Spearman Rank Correlation: {corr:.2f}')
```
