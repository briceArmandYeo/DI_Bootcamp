# Exercice 1

La visualisation des données est une étape cruciale dans l’analyse des données, car elle nous permet de traduire les données brutes en représentations visuelles plus faciles à interpréter et à comprendre. Grâce à des visualisations, nous pouvons identifier les modèles, les tendances et les idées qui pourraient ne pas être immédiatement apparents à partir de données brutes.

La visualisation des données ne consiste pas seulement à créer de jolis graphiques, mais aussi à rendre les données plus faciles à comprendre et à interpréter. Il aide à identifier les modèles, les tendances et les anomalies, rendant les données complexes plus accessibles.

#### Objectif d'un graphique linéaire dans la visualisation de données.

Les tracés de ligne sont utilisés pour afficher des valeurs quantitatives sur un intervalle continu ou une période de temps. Ils sont idéaux pour montrer les tendances et les changements dans le temps.



# Exercice 2

```python
import matplotlib.pyplot as plt

jours = ['Lundi', 'Mardi', 'Mercredi', 'Jeudi', 'Vendredi', 'Samedi', 'Dimanche']
temperatures = [72, 74, 76, 80, 82, 78, 75]

plt.plot(jours, temperatures, color='blue')

plt.xlabel('Jour')
plt.ylabel('Température (°F)')
plt.title('Évolution de la Température Hebdomadaire')
plt.show()
```

# Exercice 3

```python
month = ['Janvier', 'Fevrier', 'Mars', 'Avril', 'Mai']
sales = [5000, 5500, 6200, 7000, 7500]

plt.plot(month, sales, color='blue')

plt.xlabel('Mois')
plt.ylabel('Montant de vente ($)')
plt.title('Évolution du Montant des ventes de Janvier à Mai')
plt.show()
```

# Exercie 4

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/Student Mental health.csv')
data.head()
```

```python
print(data.isnull().sum())
```

```python
from sklearn.impute import KNNImputer
import numpy as np

imputer = KNNImputer(n_neighbors=5)

data['Age'] = imputer.fit_transform(data[['Age']])

print("Nombre de valeurs manquantes dans 'Age' après imputation :", data['Age'].isnull().sum())
data.head()
```

```python
sns.histplot(data=data, x='What is your CGPA?', color='skyblue')

plt.title('Distribution des catégories de CGPA')
plt.xlabel('Catégorie de CGPA')
plt.ylabel('Nombre d\'étudiants')
plt.show()
```

# Exercice 5

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(data=data, x='Choose your gender', hue='Do you have Anxiety?', palette='viridis')
plt.title('Comparaison de l\'anxiété par sexe chez les étudiants')
plt.xlabel('Sexe')
plt.ylabel('Nombre d\'étudiants')
plt.legend(title='A de l\'anxiété ?')

plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['Do you have Panic attack?'] = le.fit_transform(data['Do you have Panic attack?'])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(data=data, x='Age', y='Do you have Panic attack?')
plt.title('Graphique de la Relation entre l\'Age et les Attaques de Panique')
plt.xlabel('Age')
plt.ylabel('Attaque de Panique')
plt.show()
```
