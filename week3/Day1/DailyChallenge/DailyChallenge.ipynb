```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tempv1 = np.random.uniform(-5, 35, 12)
tempv2 = np.random.uniform(-5, 35, 12)
tempv3 = np.random.uniform(-5, 35, 12)
tempv4 = np.random.uniform(-5, 35, 12)
tempv5 = np.random.uniform(-5, 35, 12)
tempv6 = np.random.uniform(-5, 35, 12)
tempv7 = np.random.uniform(-5, 35, 12)
tempv8 = np.random.uniform(-5, 35, 12)
tempv9 = np.random.uniform(-5, 35, 12)
tempv10 = np.random.uniform(-5, 35, 12)
```

```python
datatemp = pd.DataFrame({'Abidjan': tempv1, 'Yamoussoukro': tempv2, 'Bouaké': tempv3, 'Daloa': tempv3,'Man': tempv4, 'Aboisso': tempv5, 'Korhogo': tempv6, 'Bouna': tempv7, 'Odiéné': tempv8, 'Bondoukou': tempv9, 'Gagnoa': tempv10 })
```

```python
moyenne_temp_ville = datatemp.mean().sort_values()

# Récupérer la première et la dernière ville sans idxmax/idxmin
ville_min = moyenne_temp_ville.index[0]
ville_max = moyenne_temp_ville.index[10]

print(f"La ville avec la température moyenne la plus élevée est : {ville_max}")
print(f"La ville avec la température moyenne la plus basse est : {ville_min}")
```

Visualisation

```python
mois = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
ville = datatemp.columns

df = pd.DataFrame(datatemp.values, index=mois, columns=datatemp.columns)


df_long = df.reset_index().rename(columns={'index': 'mois'}).melt(id_vars='mois', var_name='ville', value_name='Temperature')


palette = sns.color_palette("viridis", 11)


plt.figure(figsize=(12, 8))

sns.lineplot(
    data=df_long,
    x='mois',
    y='Temperature',
    hue='ville',
    palette=palette,
    linewidth=2.5,
    marker='o'
)

plt.title("Evolution des températures par ville")

plt.xlabel("mois")
plt.ylabel("Temperature (°C)")

plt.legend(
    title='Villes',
    bbox_to_anchor=(1.02, 1),
    loc='upper left',
    borderaxespad=0
)

plt.tight_layout()
plt.show()
```
