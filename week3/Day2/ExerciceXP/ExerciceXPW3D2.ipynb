# Exercice 6

1. Contexte

On possède les scores de productivité de 30 employés :

Avant la formation
Après la formation

L'objectif est de savoir si la formation a réellement amélioré la productivité.

2. Formuler les hypothèses

En statistique, on commence toujours par deux hypothèses.

Hypothèse nulle (H₀)

La formation n'a aucun effet.

Autrement dit : La moyenne de productivité après la formation est égale à celle d'avant.

Hypothèse alternative (H₁)
La formation augmente la productivité.



```python
np.random.seed(96)
# Generation des données
productivity_before = np.random.normal(50, 10, 30)
productivity_after = productivity_before + np.random.normal(5, 3, 30)
# Moyenne
mean_before = np.mean(productivity_before)
mean_after = np.mean(productivity_after)
# Mediane
median_before = np.median(productivity_before)
median_after = np.median(productivity_after)
# Ecart type
std_before = np.std(productivity_before)
std_after = np.std(productivity_after)
# comparaison
difference = productivity_after - productivity_before
mean_difference = np.mean(difference)
improved = np.sum(difference > 0)
print(f"Mean Before : {mean_before:.2f}")
print(f"Mean After  : {mean_after:.2f}")
print()
print(f"Median Before : {median_before:.2f}")
print(f"Median After  : {median_after:.2f}")
print()
print(f"Std Before : {std_before:.2f}")
print(f"Std After  : {std_after:.2f}")
print()
print(f"La productivité : {mean_difference:.2f}")
print("Cela signifie que la formation augmente la productivité d'environ 4 points.")

print(f"Improved Employees : {improved}")
print(f"Sur 30 employés : ont {improved} progressé")
print()

# Hypothesis conclusion
if mean_after > mean_before:
    print("\nConclusion:")
    print("Reject H0 (Hypothèse nulle).")
    print("Accept H1 (Hypothèse alternative).")
    print("La formation semble améliorer la productivité des employés.")
else:
    print("\nConclusion:")
    print("Fail to reject H0.")
    print("Il n'y a pas suffisamment d'éléments pour conclure que la formation améliore la productivité.")
```
